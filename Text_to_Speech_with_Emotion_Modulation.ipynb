```python
# Optimized_Text_to_Speech_with_Emotion_Modulation.ipynb

from gtts import gTTS
from pydub import AudioSegment

def generate_speech(text, filename, language='de', gender='female', age='adult', emotion='neutral'):
    # Generate initial speech with gTTS
    tts = gTTS(text=text, lang=language, slow=False)
    tts.save('temp.mp3')

    # Load the audio file with pydub
    sound = AudioSegment.from_mp3('temp.mp3')

    # Apply gender modulation (basic simulation)
    if gender == 'male':
        sound = sound.low_pass_filter(300).high_pass_filter(3000)

    # Apply age modulation
    if age == 'child':
        sound = sound.high_pass_filter(400).speedup(playback_speed=1.2)
    elif age == 'elderly':
        sound = sound.low_pass_filter(400).speedup(playback_speed=0.9)

    # Apply emotion modulation
    if emotion == 'curiosity':
        sound = sound.speedup(playback_speed=1.1)
    elif emotion == 'fear':
        sound = sound.speedup(playback_speed=0.9)
    elif emotion == 'tension':
        sound = sound.speedup(playback_speed=1.05)
    elif emotion == 'calm':
        sound = sound.speedup(playback_speed=0.95)
    elif emotion == 'happiness':
        sound = sound.speedup(playback_speed=1.15)
    elif emotion == 'sadness':
        sound = sound.speedup(playback_speed=0.85)

    # Export the modified sound
    sound.export(filename, format="mp3")
    print(f"Speech generated and saved as {filename}")

# Beispiele
text_1 = "Hallo, wie geht es Ihnen heute? Ich hoffe, Sie haben einen großartigen Tag."
generate_speech(text_1, 'output_curiosity.mp3', language='de', gender='female', age='adult', emotion='curiosity')

text_2 = "Oh nein, was ist das für ein Geräusch? Ich habe Angst."
generate_speech(text_2, 'output_fear.mp3', language='de', gender='male', age='adult', emotion='fear')

text_3 = "Spannung liegt in der Luft, was wird als nächstes passieren?"
generate_speech(text_3, 'output_tension.mp3', language='de', gender='female', age='elderly', emotion='tension')

text_4 = "Alles wird gut, bleiben Sie ruhig und entspannen Sie sich."
generate_speech(text_4, 'output_calm.mp3', language='de', gender='male', age='adult', emotion='calm')

text_5 = "Hurra! Wir haben gewonnen, das ist fantastisch!"
generate_speech(text_5, 'output_happiness.mp3', language='de', gender='female', age='child', emotion='happiness')

text_6 = "Es tut mir so leid, ich fühle mich sehr traurig."
generate_speech(text_6, 'output_sadness.mp3', language='de', gender='female', age='elderly', emotion='sadness')
```
